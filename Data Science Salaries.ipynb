# Data Preparation:


### 1. Write a python program to load data into pandas DataFrame 

```python
# Importing pandas for pannel data analysis and aliasing it as pd.
import pandas as pd

# Importing matplotlib.pyplot and aliasing it as plt.
import matplotlib.pyplot as plt
```

```python
# Declaring dataset as variable and storing retrived data set from CSV file.
dataset = pd.read_csv('DataScienceSalaries.csv')

# Printing the declared variable.
dataset
```

### 2. Write a python program to remove unnecessary columns i.e., salary and salary currency.

```python
# Removing the columns salary and salary_currency from the dataset DataFrame.
dataset.drop(columns=['salary','salary_currency'],inplace=True)

# Displaying the valuehs after removing columns.
dataset
```

### 3. Write a python program to remove the NaN missing values from updated dataframe. 

```python
# Removing the NaN(null value) from the dataset by using the dropna method.
RemoveNaNFromDF = dataset.dropna()

# Printing the values after removing empty data from row.
RemoveNaNFromDF
```

### 4. Write a python program to check duplicates value in the dataframe

```python
# Checking for duplicated rows in the dataset by using the duplicated method.
DuplicatedValues = dataset.duplicated(keep='first')

# Listing the checked data and displaying it by using data frame method
pd.DataFrame({'Duplicated':DuplicatedValues.values})
```

```python
'''
Checking for duplicated rows in the dataset by using the duplicated method.
Listing the first occurence of the duplicated data by the keep parameters
'''
DuplicatedValues = dataset[dataset.duplicated(keep='first')]

# Displaying the duplicated valued which are occured at first
DuplicatedValues
```

### 5. Write a python program to see the unique values from all the columns in the dataframe.

```python
# Iterating through each column in the dataset.
for column in dataset:
    
    # Printing  message with column name.
    print("\n Unique Values in", column)
    
    # Printing unique values of each column.
    print(dataset[column].unique())
    
```

# Data Analysis

### 1. Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable. 

```python
# Calculating the sum of USD salaries of the dataset using the sum method.
Sum_of_UsdSalary = dataset['salary_in_usd'].sum()

# Calculating the mean of USD salaries of the dataset using the mean method.
Mean_of_UsdSalary = dataset['salary_in_usd'].mean()

# Calculating the standard deviation of USD salaries of the dataset using the std method.
Deviation_of_UsdSalary = dataset['salary_in_usd'].std()

# Calculating the skewness of USD salaries of the dataset using the skew method.
Skewness_of_UsdSalary = dataset['salary_in_usd'].skew()

# Calculating the kurtosis of USD salaries of the dataset using the sum method.
Kurtosis_of_UsdSalary = dataset['salary_in_usd'].kurt()

# Displaying all the above calculation
print ('The sum of the USD salary =',Sum_of_UsdSalary)
print ('The mean of USD salary =',Mean_of_UsdSalary)
print ('The standard deviation of USD salary =',Deviation_of_UsdSalary)
print ('The skewness of USD salary =',Skewness_of_UsdSalary)
print ('The kurtosis of USD salary =',Kurtosis_of_UsdSalary)
```

# Data Exploration

### 1. Write a python program to find out top 15 jobs. Make a bar graph of salary as well.         	         

```python
# Grouping the job title by group method and aggregating
TopJobs = dataset.groupby('job_title').agg(count=('salary_in_usd', 'size'), mean_salary=('salary_in_usd', 'mean'))

# Sorting by mean_salary and count
TopJobs = TopJobs.sort_values(by='mean_salary', ascending = False)
TopJobs = TopJobs.sort_values(by='count', ascending = False).head(15)

# Seting the figure size
plt.figure(figsize=(10,6))

# Plotting the bars
plt.bar(TopJobs.index, TopJobs['count'], color = '#C6C6C6')

# Setting background color
plt.gca().set_facecolor('#F2F2F2')

# Adding the labels and title
plt.xlabel('Job Title', fontsize=18, fontweight='bold', color='#005A34')
plt.ylabel('Occurrence', fontsize=18, fontweight='bold', color='#005A34')
plt.title('Top Jobs by Occurrence', fontsize=20, fontweight='bold', color='#005A34')

# Rotate x-axis labels for better readability
plt.xticks(rotation = -90, color = '#324B4C', fontsize = 12)

# Showing the plot
plt.show()
```

```python
# Setting the figure size.
plt.figure(figsize=(10,6))

# Plotting the bargraph for  top 15 jobs by avg salary.
plt.bar(TopJobs.index, TopJobs['mean_salary'], color = '#C6C6C6')

# Setting the background color.
plt.gca().set_facecolor('#F2F2F2')

# Adding the labels and title
plt.xlabel('Job Title', fontsize=18, fontweight='bold', color='#005A34')
plt.ylabel('Average Salary', fontsize=18, fontweight='bold', color='#005A34')
plt.title('Top Jobs Based on Average Salary', fontsize=20, fontweight='bold', color='#005A34')

# Rotate x-axis labels for better readability
plt.xticks(rotation= -90, color = '#324B4C', fontsize = 12)

# Displaying the plot
plt.show()
```

### 2. Which job has the highest salaries? Illustrate with bar graph.		

```python
# Setting the figure size.
plt.figure(figsize=(10,6))

# Grouping the dataset by job_title and calculating the mean salary for each job title.
highest_salaries_jobs = dataset.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(15).plot(kind='bar', color = '#C6C6C6')

# Setting the background color.
plt.gca().set_facecolor('#F2F2F2')

# Adding the labels and title
plt.xlabel('Job Title', fontsize=18, fontweight='bold', color='#005A34')
plt.ylabel('Average Salary', fontsize=18, fontweight='bold', color='#005A34')
plt.title('Highest Salaries Jobs', fontsize=20, fontweight='bold', color='#005A34')

# Rotating x-axis labels for better readability.
plt.xticks(rotation = -90, color = '#324B4C', fontsize = 12) 

# Displaying the bar graph.
plt.show()
```

### 3. Write a python program to find out salaries based on experience level. Illustrate it through bar graph.

```python
# Grouping the dataset by experience_level and calculating the mean salary for each experience level.
SalariesBasedExperience = dataset.groupby('experience_level')['salary_in_usd'].mean()

# Setting the figure size.
plt.figure(figsize=(10,6))

# Plotting the bargraphs.
SalariesBasedExperience.plot(kind='bar', color = '#C6C6C6')

# Setting the background color.
plt.gca().set_facecolor('#F2F2F2')

# Adding the labels and title
plt.xlabel('Experience', fontsize=18, fontweight='bold', color='#005A34')
plt.ylabel('Average Salary', fontsize=18, fontweight='bold', color='#005A34')
plt.xticks(rotation = -90, color = '#324B4C', fontsize = 12)
plt.title('Average Salary Based on Experience Level', fontsize=20, fontweight='bold', color='#005A34')

# Displaying the plot
plt.show()
```
